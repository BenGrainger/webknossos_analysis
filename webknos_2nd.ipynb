```python
new_dataset_name='B002_04x'

dataset = wk.Dataset.download(
        dataset_name='B002_uCT_04x',
        organization_name="Schaefer_Lab",
        path=new_dataset_name,
        webknossos_url="https://webknossos.crick.ac.uk",
)
```

```python
dataset = wk.Dataset.open('B002_4x')
```

```python
dataset.layers
```

```python
layer = dataset.get_layer('versaCT')
layer
```

```python
mag1 = layer.get_mag("1")
```

```python
data_array = mag1.read()
```

```python
data_array[0].shape
```

```python
data_array[0][0].shape
```

```python
# 500 pixels = 2mm
save_loc = 'B002_4x.svg'
fig, ax = plt.subplots(1,1, figsize=(10,10))
plt.imshow(data_array[0][370][200:], cmap='Greys')
ax.set_yticks([])
ax.set_xticks([])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
plt.plot([700, 950], [760, 760], 'k-', lw=5)
fig.savefig(save_loc)
```

```python
dataset2 = wk.Dataset.open('B002_04x')
layer2 = dataset2.get_layer('versaCT')
mag04 = layer2.get_mag("1")
_04_data = mag04.read()
```
